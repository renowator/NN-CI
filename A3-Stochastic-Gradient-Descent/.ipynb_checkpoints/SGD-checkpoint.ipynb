```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ksi = np.genfromtxt('xi.csv', delimiter=',')
tau = np.genfromtxt('tau.csv', delimiter=',')
```

```python
# Makes a random vector with dims dimensions and length 1
def make_rand_vector(dims):
    vec = [random.gauss(-1, 1) for i in range(dims)]
    mag = sum(x**2 for x in vec) ** .5
    return [x/mag for x in vec]
```

```python
class SoftCommitteeMachine:
    
    def __init__(self, features, labels, P = 100, Q = 100, vk = 1, K = 2, t_max = 10):
        self.debug = False
        self.ksi = features.T
        self.tau = labels
        self.vk = vk
        self.K = K
        if P + Q  <= len(self.tau[:]):
            self.P = P
            self.Q = Q
        else:
            self.P = 0
            self.Q = 0
        self.N = len(self.ksi[:][0])
        self.nu = 0.05
        self.W = []
        for i in range(self.K):
            self.W.append(make_rand_vector(self.N))
        self.t_max = t_max
        self.t = 0
        self.E = 0
        self.E_test = 0


    
    def stochastic_gradient_update(self):
        choice = random.randint(0,self.P-1)
        ksi_v = self.ksi[choice]
        tau_v = self.tau[choice]
        sigma_v = 0
        for i in range(self.K):
            sigma_v += self.vk*np.tanh(np.dot(self.W[i], ksi_v))
        diff = sigma_v - tau_v
        e_v = (diff*diff)/2
        if self.debug:
            print("Item error:", e_v)
        # Gradient equations:
        # d(e_v)/d w_k =((sigma(w dot ksi)-tao)^2/2)' = (sigma(w dot ksi)-tao)(sigma(w dot ksi)-tao)'
        # (sigma(w dot ksi - tao))' = (tanh(w_k dot ksi))' = (1 - tanh^2(w_k dot ksi))*ksi
        dw1 = diff*(1 - np.tanh(np.dot(self.W[0], ksi_v))**2)*ksi_v
        dw2 = diff*(1 - np.tanh(np.dot(self.W[1], ksi_v))**2)*ksi_v
        if self.debug:
            print(dw1)
            print(dw2)
        self.W[0] = self.W[0] - self.nu*dw1*e_v
        self.W[1] = self.W[1] - self.nu*dw2*e_v
        
    def calculate_train_error(self):
        diff = 0
        for j in range(self.P):
            sigma = 0
            for i in range(self.K):
                sigma += self.vk*np.tanh(np.dot(self.W[i], self.ksi[j]))
            diff += ((sigma - self.tau[j])*(sigma - self.tau[j]))/2
        self.E = diff/self.P
        if self.debug:
            print("Training Dataset error:", self.E)
        return self.E
    
    def calculate_test_error(self):
        diff = 0
        for j in range(self.Q):
            sigma = 0
            for i in range(self.K):
                sigma += self.vk*np.tanh(np.dot(self.W[i], self.ksi[self.P + j]))
            diff += ((sigma - self.tau[self.P+j])*(sigma - self.tau[self.P+j]))/2
        self.E_test = diff/self.Q
        if self.debug:
            print("Testing Dataset error:", self.E_test)
        return self.E_test
    
    def plot_weights(self):
        weights = []
        i = 1
        plt.figure(figsize=(30,10))
        for item in self.W[0]:
            weights.append(str(i))
            i += 1
        plt.bar(weights, self.W[0], color = 'b', label="W1 weights", width=0.5)
        plt.bar(weights, self.W[1], color = 'r', label="W2 weights", width=0.5)
        plt.legend( shadow=True, fontsize='large')
        plt.show()
        
    def train_network_and_display_losses(self):
        e_train = []
        e_test = []
        while self.t <= self.t_max*self.P:
            #print("Gradient update Iteration ", a.t)
            self.stochastic_gradient_update()
            if self.t % self.P == 0:
                e_test.append(self.calculate_test_error())
                e_train.append(self.calculate_train_error())
            self.t += 1
        plt.figure(figsize=(10,5))
        plt.plot(e_train, label= "Training error")
        plt.plot(e_test, label= "Testing error")
        plt.legend( shadow=True, fontsize='large')
        plt.show()
```

```python
P_Q_values = [50, 100, 200, 500, 1000, 2000]
for value in P_Q_values:
    print("P and Q values are set to " + str(value))
    machine = SoftCommitteeMachine(ksi, tau, P=value, Q=value, t_max=5)
    machine.train_network_and_display_losses()
    machine.plot_weights()
```
